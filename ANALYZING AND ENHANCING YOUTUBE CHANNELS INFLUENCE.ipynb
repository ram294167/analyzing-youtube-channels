# function for extraction of channel statistics


```python
def get_channel_stats(youtube,channels_list):
    channels_data=[]
    request=youtube.channels().list(
    part='snippet,contentDetails,statistics',id=channels_list)
    response=request.execute()
    for i in range(len(response['items'])):
        data=dict(
        Channel_name=response['items'][i]['snippet']['title'],
        Channel_id=response['items'][i]['id'],
        No_of_Subscribers=response['items'][i]['statistics']['subscriberCount'],
        No_of_Views=response['items'][i]['statistics']['viewCount'],
        No_of_Videos=response['items'][i]['statistics']['videoCount'],
        Published_date=response['items'][i]['snippet']['publishedAt'],
        Playlist_id=response['items'][i]['contentDetails']['relatedPlaylists']['uploads'])
        channels_data.append(data)
    return channels_data
```

```python
def get_video_ids(youtube,playlist_id):
    request=youtube.playlistItems().list(
            part="contentDetails",
            playlistId=playlist_id,
            maxResults=50)
    response=request.execute()
    video_ids=[];
    for i in range(len(response['items'])):
        video_ids.append(response['items'][i]['contentDetails']['videoId'])
    next_page_token=response.get('nextPageToken')
    more_pages=True
    while(more_pages):
        if next_page_token is None:
            more_pages=False
        else:
            request=youtube.playlistItems().list(
            part="contentDetails",
            playlistId=playlist_id,
            maxResults=50,
            pageToken=next_page_token)
            response=request.execute()
            for i in range(len(response['items'])):
                video_ids.append(response['items'][i]['contentDetails']['videoId']);
            next_page_token=response.get('nextPageToken')
    return video_ids
```

## Function for playlist statistics

```python
import re
from datetime import timedelta

def durations(li):
    api_key='AIzaSyB6TWw8weJykuk7JkBHmeT6k22zzv5zU4I'
    youtube=build('youtube','v3',developerKey=api_key)
    final_dic={}
    for playlistid in li:
        list1=[]
        total_seconds=0
        views=0;
        likes=0;
        hours_pattern=re.compile(r'(\d+)H')
        minutes_pattern=re.compile(r'(\d+)M')
        seconds_pattern=re.compile(r'(\d+)S')
        nextPageToken=None
        while True:
            pl_request=youtube.playlistItems().list(
            part='contentDetails,snippet',
            playlistId=playlistid,
            maxResults=50,
            pageToken=nextPageToken)
            pl_response=pl_request.execute()
            vid_ids=[]
            for item in pl_response['items']:
                vid_ids.append(item['contentDetails']['videoId'])
                playlist_name=item['snippet']['title']
            list1.append(playlist_name)
            list1.append(len(vid_ids))
            vid_request=youtube.videos().list(
            part='contentDetails,statistics',
            id=','.join(vid_ids))
            vid_response=vid_request.execute()
            for item in vid_response['items']:
                duration=item['contentDetails']['duration']
                if item['statistics'].get('viewCount')!=None:
                    views+=int(item['statistics']['viewCount']);
                if(item['statistics'].get('likeCount')!=None):
                    likes+=int(item['statistics']['likeCount']);
                hours=hours_pattern.search(duration)
                minutes=minutes_pattern.search(duration)
                seconds=seconds_pattern.search(duration)
                hours=int(hours.group(1)) if(hours) else 0
                minutes=int(minutes.group(1)) if(minutes) else 0
                seconds=int(seconds.group(1)) if(seconds) else 0
                video_seconds=timedelta(hours=hours,
                                       minutes=minutes,
                                       seconds=seconds).total_seconds()
                total_seconds+=video_seconds
            list1.append(total_seconds)
            list1.append(views);
            list1.append(likes);
            minutes,seconds=divmod(total_seconds,60)
            hours,minutes=divmod(minutes,60)
            duration=str(int(hours))+":"+str(int(minutes))+":"+str(int(seconds))
            list1.append(duration)
            final_dic[playlistid]=list1
            nextPageToken=pl_response.get("nextPageToken")
            if not nextPageToken:
                break
    return(final_dic)
```

## Creating the dataframe of playlist statistics

```python

def CreateDataframe(final_dic):
    labels=[]
    total_videos=[]
    total_seconds=[]
    views=[]
    likes=[]
    duration=[]
    for key,value in final_dic.items():
        labels.append(value[0])
        total_videos.append(value[1])
        total_seconds.append(value[2])
        views.append(value[3])
        likes.append(value[4])
        duration.append(value[5])
    playlist_details=pd.DataFrame(columns=['playlist_name','total_videos','total_seconds','duration'])
    playlist_details['playlist_name']=labels
    playlist_details['total_videos']=total_videos
    playlist_details['total_seconds']=total_seconds
    playlist_details['total_views']=views
    playlist_details['total_likes']=likes
    playlist_details['duration']=duration
    return playlist_details
```

```python
import pandas as pd
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
import time
import logging


logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)


API_KEY = 'AIzaSyCjiE9R3Zn3wMGpz1Xttq-quHfC__mICb4'
youtube = build('youtube', 'v3', developerKey=API_KEY)

def get_video_data(youtube, playlist_id, max_videos=1):
    video_data = []
    next_page_token = None
    retries = 0
    max_retries = 5
    videos_fetched = 0

    playlist_title = ''
    channel_name = ''
    try:
        playlist_request = youtube.playlists().list(
            part='snippet',
            id=playlist_id
        )
        playlist_response = playlist_request.execute()
        playlist_title = playlist_response['items'][0]['snippet']['title']
        channel_name = playlist_response['items'][0]['snippet']['channelTitle']
    except HttpError as e:
        logger.error(f"HTTP error occurred while fetching playlist title: {e}")

    while True:
        try:
            request = youtube.playlistItems().list(
                part='snippet',
                playlistId=playlist_id,
                maxResults=50,
                pageToken=next_page_token
            )
            response = request.execute()

            for item in response['items']:
                if videos_fetched >= max_videos:
                    break
                video_id = item['snippet']['resourceId']['videoId']
                video_request = youtube.videos().list(
                    part='snippet',
                    id=video_id
                )
                video_response = video_request.execute()
                for video_item in video_response['items']:
                    video_tags = video_item['snippet'].get('tags', [])
                    video_data.append({'video_id': video_id, 'tags': video_tags})
                    videos_fetched += 1

            next_page_token = response.get('nextPageToken')
            if not next_page_token or videos_fetched >= max_videos:
                break

        except HttpError as e:
            if e.resp.status == 403 and 'quotaExceeded' in str(e):
                retries += 1
                if retries > max_retries:
                    logger.error("Max retries exceeded. Quota still exceeded.")
                    return video_data, playlist_title, channel_name  # Return what we have so far
                sleep_time = 2 ** retries * 60  # Exponential backoff
                logger.warning(f"Quota exceeded. Sleeping for {sleep_time / 60} minutes.")
                time.sleep(sleep_time)
            else:
                logger.error(f"HTTP error occurred: {e}")
                break

    return video_data, playlist_title, channel_name


def categorize_channel(video_data):
    categories = {
        'Education': ['tutorial', 'how to', 'lesson', 'learn', 'education', 'study', 'course'],
        'Entertainment': ['funny', 'entertainment', 'comedy', 'vlog', 'music', 'show', 'series'],
        'Technology': ['tech', 'gadget', 'review', 'smartphone', 'computer', 'software'],
        'Gaming': ['game', 'gaming', 'gamer', 'let\'s play', 'walkthrough'],
        'Health & Fitness': ['health', 'fitness', 'workout', 'exercise', 'nutrition', 'diet'],
        'Food & Cooking': ['recipe', 'cooking', 'food', 'baking', 'kitchen', 'cuisine'],
        'Beauty & Fashion': ['beauty', 'makeup', 'fashion', 'style', 'skincare'],
        'News': ['news', 'report', 'journalism', 'current events', 'politics'],
        'Travel': ['travel', 'trip', 'destination', 'tour', 'vacation'],
        'Sports': ['sports', 'soccer', 'basketball', 'football', 'baseball', 'athlete'],

    }

    tag_count = {category: 0 for category in categories}

    for video in video_data:
        for tag in video['tags']:
            for category, keywords in categories.items():
                if any(keyword in tag.lower() for keyword in keywords):
                    tag_count[category] += 1

    return max(tag_count, key=tag_count.get) if tag_count else 'Unknown'


def categorize_channels(youtube, channels_data):
    categorized_channels = []

    for channel_data in channels_data:
        try:
            channel_id = channel_data['Channel_id']
            playlist_id = channel_data['Playlist_id']

            video_data, playlist_title, channel_name = get_video_data(youtube, playlist_id, max_videos=1)
            category = categorize_channel(video_data)

            categorized_channels.append({
                'Channel_id': channel_id,
                'Channel_name': channel_name,
                'Playlist_id': playlist_id,
                'Playlist_title': playlist_title,
                'Category': category
            })

        except HttpError as e:
            logger.error(f"HTTP error occurred while processing channel {channel_id}: {e}")
        except Exception as e:
            logger.error(f"An unexpected error occurred while processing channel {channel_id}: {e}")

        time.sleep(2)

    return pd.DataFrame(categorized_channels)

channel_playlist_data = [
    {'Channel_id': 'UCVDGr6BSFx9wneWkpQRHhNA', 'Playlist_id': 'UUVDGr6BSFx9wneWkpQRHhNA'},
    {'Channel_id': 'UCt4atlExw8aj3Bm79nv1fig', 'Playlist_id': 'UUt4atlExw8aj3Bm79nv1fig'},
    {'Channel_id': 'UCwD6wSolY9DYNGvR9KgHN-w', 'Playlist_id': 'UUwD6wSolY9DYNGvR9KgHN-w'},
    {'Channel_id': 'UC0C-w0YjGpqDXGB8IHb662A', 'Playlist_id': 'UU0C-w0YjGpqDXGB8IHb662A'},
    {'Channel_id': 'UCqECaJ8Gagnn7YCbPEzWH6g', 'Playlist_id': 'UUqECaJ8Gagnn7YCbPEzWH6g'},
    {'Channel_id': 'UCP255IH-9gfje74L4r5ktYQ', 'Playlist_id': 'UUP255IH-9gfje74L4r5ktYQ'},
    {'Channel_id': 'UCKS34cSMNaXaySe2xgXH-3A', 'Playlist_id': 'UUKS34cSMNaXaySe2xgXH-3A'},
    {'Channel_id': 'UC9CoOnJkIBMdeijd9qYoT_g', 'Playlist_id': 'UU9CoOnJkIBMdeijd9qYoT_g'},
    {'Channel_id': 'UC63URkuUvnugRBeTNqmToKg', 'Playlist_id': 'UU63URkuUvnugRBeTNqmToKg'},
    {'Channel_id': 'UCJihyK0A38SZ6SdJirEdIOw', 'Playlist_id': 'UUJihyK0A38SZ6SdJirEdIOw'},
    {'Channel_id': 'UCd0L83mMeQ4Ioy4IjaSYhvA', 'Playlist_id': 'UUd0L83mMeQ4Ioy4IjaSYhvA'},
    {'Channel_id': 'UUq-Fj5jknLsUf-MWSy4_brA', 'Playlist_id': 'UUq-Fj5jknLsUf-MWSy4_brA'},
    {'Channel_id': 'UCDxNoaHOj2l2bFrHgCkg6Zw', 'Playlist_id': 'UUDxNoaHOj2l2bFrHgCkg6Zw'},
    {'Channel_id': 'UU0RhatS1pyxInC00YKjjBqQ', 'Playlist_id': 'UU0RhatS1pyxInC00YKjjBqQ'},
    {'Channel_id': 'UUiGm_E4ZwYSHV3bcW1pnSeQ', 'Playlist_id': 'UUiGm_E4ZwYSHV3bcW1pnSeQ'},
    {'Channel_id': 'UUN02qVJEtVzieJWZnSKv7QA', 'Playlist_id': 'UUN02qVJEtVzieJWZnSKv7QA'},
]
categorized_data = categorize_channels(youtube, channel_playlist_data)


print(categorized_data)
```

```python
from googleapiclient.discovery import build
import pandas as pd
import time
import logging
from langdetect import detect, LangDetectException
from googletrans import Translator


logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)


API_KEY = 'AIzaSyAjlevt6Hi0LZRMDDX5nWgWBFpOGw3mNto'
youtube = build('youtube', 'v3', developerKey=API_KEY)


translator = Translator()


def get_video_comments(youtube, video_id, max_results=100):
    comments = []
    next_page_token = None

    while True:
        try:

            response = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                maxResults=min(max_results, 100),
                pageToken=next_page_token
            ).execute()


            for item in response['items']:
                comment = item['snippet']['topLevelComment']['snippet']['textDisplay']

                try:
                    if detect(comment) != 'en':
                        translated_comment = translator.translate(comment, dest='en').text
                        comments.append(translated_comment)
                    else:
                        comments.append(comment)
                except LangDetectException:
                    continue


            next_page_token = response.get('nextPageToken')
            if not next_page_token:
                break

            time.sleep(1)

        except Exception as e:
            logger.error(f"Error fetching comments for video {video_id}: {e}")
            break

    return comments


def get_uploads_playlist_id(youtube, channel_id):
    try:
        response = youtube.channels().list(
            part='contentDetails',
            id=channel_id
        ).execute()

        uploads_playlist_id = response['items'][0]['contentDetails']['relatedPlaylists']['uploads']
        return uploads_playlist_id

    except Exception as e:
        logger.error(f"Error fetching uploads playlist ID for channel {channel_id}: {e}")
        return None


def get_playlist_comments(youtube, playlist_id, channel_name, max_videos=3):
    playlist_comments = []
    videos_retrieved = 0

    try:

        next_page_token = None
        while videos_retrieved < max_videos:
            playlist_request = youtube.playlistItems().list(
                part='snippet',
                playlistId=playlist_id,
                maxResults=50,
                pageToken=next_page_token
            )
            playlist_response = playlist_request.execute()

            for video_item in playlist_response['items']:
                if videos_retrieved >= max_videos:
                    break
                video_id = video_item['snippet']['resourceId']['videoId']
                video_title = video_item['snippet']['title']

                video_comments = get_video_comments(youtube, video_id)


                for comment in video_comments:
                    playlist_comments.append({
                        'Channel_name': channel_name,
                        'Video_title': video_title,
                        'Comment': comment
                    })

                videos_retrieved += 1

            next_page_token = playlist_response.get('nextPageToken')
            if not next_page_token:
                break

            time.sleep(2)

    except Exception as e:
        logger.error(f"Error processing playlist {playlist_id} for channel {channel_name}: {e}")

    return playlist_comments

def get_channel_comments(youtube, channel_data):
    channel_comments = []

    for channel in channel_data:
        try:
            channel_name = channel['Channel_name']
            channel_id = channel['Channel_id']


            playlist_id = get_uploads_playlist_id(youtube, channel_id)
            if not playlist_id:
                continue


            playlist_comments = get_playlist_comments(youtube, playlist_id, channel_name)
            channel_comments.extend(playlist_comments)

        except Exception as e:
            logger.error(f"Error processing channel {channel_name}: {e}")

    return pd.DataFrame(channel_comments)


if __name__ == "__main__":
    channel_data = [
      {'Channel_name': 'Gate Smashers', 'Channel_id': 'UCJihyK0A38SZ6SdJirEdIOw'},
    {'Channel_name': 'Dinesh Varyani', 'Channel_id': 'UCd0L83mMeQ4Ioy4IjaSYhvA'},
    {'Channel_name': 'T-Series', 'Channel_id': 'UCq-Fj5jknLsUf-MWSy4_brA'},
    {'Channel_name': 'Sudhakar Atchala', 'Channel_id': 'UCDxNoaHOj2l2bFrHgCkg6Zw'},
    {'Channel_name': 'GeeksforGeeks', 'Channel_id': 'UC0RhatS1pyxInC00YKjjBqQ'},
    {'Channel_name': 'Billie Eilish', 'Channel_id': 'UCiGm_E4ZwYSHV3bcW1pnSeQ'},
    {'Channel_name': 'Anusha world', 'Channel_id': 'UCN02qVJEtVzieJWZnSKv7QA'},
    ]

    channel_comments_df = get_channel_comments(youtube, channel_data)
    print(channel_comments_df)
```

```python
import re


def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           u"\U00002500-\U00002BEF"
                           u"\U00002702-\U000027B0"
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           u"\U0001f926-\U0001f937"
                           u"\U00010000-\U0010ffff"
                           u"\u2640-\u2642"
                           u"\u2600-\u2B55"
                           u"\u200d"
                           u"\u23cf"
                           u"\u23e9"
                           u"\u231a"
                           u"\ufe0f"
                           u"\u3030"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)


channel_comments_df['Comment'] = channel_comments_df['Comment'].apply(remove_emojis)


print(channel_comments_df)
```

```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import numpy as np  # Import numpy for NaN handling

nltk.download('vader_lexicon')


df = pd.read_csv('/home/channel_comments (2).csv')


df['Comment'] = df['Comment'].fillna('')


sid = SentimentIntensityAnalyzer()

def get_sentiment_score(comment):
    scores = sid.polarity_scores(comment)
    return scores['compound']


df['Sentiment'] = df['Comment'].apply(get_sentiment_score)


def get_sentiment_label(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['Sentiment'].apply(get_sentiment_label)



```

```python

import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


df = pd.read_csv('/home/sentiment_analysis_result (1).csv')

channel_sentiment = df.groupby('Channel_name')['Sentiment'].apply(lambda x: x.mode().iloc[0]).reset_index()


X = channel_sentiment[['Channel_name']]
y = channel_sentiment['Sentiment']

preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(), ['Channel_name'])
    ],
    remainder='passthrough'
)

X_encoded = preprocessor.fit_transform(X)

rf_classifier = RandomForestClassifier(random_state=42)

rf_classifier.fit(X_encoded, y)

unique_channels = pd.DataFrame({'Channel_name': df['Channel_name'].unique()})
X_pred_encoded = preprocessor.transform(unique_channels[['Channel_name']])
predictions = rf_classifier.predict(X_pred_encoded)


predictions_df = pd.DataFrame({'Channel_name': unique_channels['Channel_name'], 'Predicted_Sentiment': predictions})

print(predictions_df)
```
